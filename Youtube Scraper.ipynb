# Importing Libraries

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
from tqdm import tqdm
```

# Generating the Youtube link

```python
browser = webdriver.Chrome()

browser.get("https://www.youtube.com/@BBKiVines/videos")
```

# Parsing the Link

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

# Extracting all the data from video section

```python
data = []

for sp in soup.find_all("div", class_= "style-scope ytd-rich-item-renderer"):

#     print(sp.text.strip())
    title = sp.find('a', class_= "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media" ).text
    video_link = sp.find('a', class_= "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media" ).get('href')
    views = sp.find_all('span', class_="inline-metadata-item style-scope ytd-video-meta-block")[0].text
    date_time = sp.find_all('span', class_="inline-metadata-item style-scope ytd-video-meta-block")[1].text
    
    try:
        thumbnail = sp.find('img', class_ ="yt-core-image yt-core-image--fill-parent-height yt-core-image--fill-parent-width yt-core-image--content-mode-scale-aspect-fill yt-core-image--loaded").get('src').split("?")[0]
    except:
        thumbnail=None
    
    print(thumbnail)
    print("%"*66)
    data.append([title,views,date_time,video_link,thumbnail])
    
```

# Saving the data into csv file

```python
df= pd.DataFrame(data, columns=['Title', 'Views','Date_Time','video_link','Thumbnail'])
df.to_csv("BB-KI-VINES_Youtube_Data.csv", index=False)
```

# NOW extract data from every particular videos 

# # Reading the data file

```python
df = pd.read_csv("BB-KI-VINES_Youtube_Data.csv")
```

# extracting imporant data from every video

```python
data = []
browser = webdriver.Chrome()   #making web drivers

browser.get("https://www.youtube.com")   # using youtube link
time.sleep(2)                             # use to have some buffer time to load the page correctly

for link in tqdm(df['video_link']):        #usinng video link retrived from video section of the youtube channel
    
    link = "https://www.youtube.com"+ link
    browser.get(link)
    
    time.sleep(5)                         # use to have some buffer time to load the page correctly
    
    sp = BeautifulSoup(browser.page_source, 'html.parser')  # parsing the page
    
    try:
        title = sp.find('yt-formatted-string',  class_ = "style-scope ytd-watch-metadata").text
    except:
        title = np.nan
        
    try:
        likes = sp.find_all('div', class_= "yt-spec-button-shape-next__button-text-content")[2].text
    except:
        likes = np.nan
        
    try:
        views = sp.find_all('span', class_= "bold style-scope yt-formatted-string")[0].text
    except:
        views= np.nan
        
    try:
        date_time = sp.find_all('span', class_= "bold style-scope yt-formatted-string")[2].text
    except:
        date_time= np.nan
        
    try:
        description = sp.find('span', class_="yt-core-attributed-string yt-core-attributed-string--white-space-pre-wrap").text
    except:
        description = np.nan
        
    data.append([title, likes, views, date_time, link, description])
```

# Saving the data

```python
df= pd.DataFrame(data, columns=['Title','Likes', 'Views','Date_Time','video_link','Description'])
df.to_csv("BB-KI-VINES_Youtube_Data_Each_Video.csv", index=False)
```
